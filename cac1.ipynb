```python
# Define a dictionary of contents with subcategories and recommendations
contents = {
    "Movies": {
        "Action Movies": ["Die Hard", "Mad Max: Fury Road", "The Dark Knight"],
        "Comedy Movies": ["Superbad", "Anchorman", "The Grand Budapest Hotel"],
        "Drama Movies": ["The Shawshank Redemption", "Forrest Gump", "The Godfather"],
        "Science Fiction Movies": ["Blade Runner", "The Matrix", "Interstellar"],
    },
    "Books": {
        "Fiction": ["To Kill a Mockingbird", "1984", "The Great Gatsby"],
        "Non-Fiction": ["Sapiens", "Becoming", "Educated"],
        "Mystery": ["Gone Girl", "The Girl with the Dragon Tattoo", "Sherlock Holmes"],
        "Science Fiction": ["Dune", "Ender's Game", "Foundation"],
        "Fantasy": ["Harry Potter", "The Lord of the Rings", "A Song of Ice and Fire"],
    },
    "Music": {
        "Rock": ["Led Zeppelin", "The Rolling Stones", "Pink Floyd"],
        "Pop": ["Michael Jackson", "Adele", "Taylor Swift"],
        "Jazz": ["Miles Davis", "John Coltrane", "Ella Fitzgerald"],
        "Classical": ["Beethoven", "Mozart", "Bach"],
    },
    "TV Shows": {
        "Drama Series": ["Breaking Bad", "Game of Thrones", "The Crown"],
        "Comedy Series": ["Friends", "The Office", "Brooklyn Nine-Nine"],
        "Crime Series": ["Sherlock", "Mindhunter", "Narcos"],
        "Fantasy Series": ["Game of Thrones", "The Witcher", "The Mandalorian"],
    },
    "Video Games": {
        "Action Games": ["Assassin's Creed", "Call of Duty", "GTA V"],
        "Adventure Games": ["The Legend of Zelda", "Uncharted", "Red Dead Redemption"],
        "RPG Games": ["The Elder Scrolls V: Skyrim", "The Witcher 3", "Final Fantasy VII"],
        "Strategy Games": ["Civilization VI", "Stellaris", "Total War: Three Kingdoms"],
    },
    "Food": {
        "Italian Cuisine": ["Pizza Margherita", "Pasta Carbonara", "Tiramisu"],
        "Mexican Cuisine": ["Tacos", "Guacamole", "Enchiladas"],
        "Chinese Cuisine": ["Kung Pao Chicken", "Dim Sum", "Peking Duck"],
        "Indian Cuisine": ["Chicken Tikka Masala", "Naan", "Biryani"],
    },
    "Travel": {
        "Beach Destinations": ["Maldives", "Bora Bora", "Hawaii"],
        "Mountain Retreats": ["Swiss Alps", "Rocky Mountains", "Himalayas"],
        "City Breaks": ["Paris", "New York City", "Tokyo"],
        "Cultural Experiences": ["Rome", "Kyoto", "Machu Picchu"],
    },
}

# Create or open a file for appending recommendations
def append_recommendations_to_file(category, subcategory, recommendations):
    filename = f"{category}_{subcategory}_recommendations.txt"
    with open(filename, "a") as file:
        for recommendation in recommendations:
            file.write(f"- {recommendation}\n")

# Display a menu for categories
def display_menu():
    print("Categories:")
    for index, category in enumerate(contents.keys(), 1):
        print(f"{index}. {category}")
    print(f"{len(contents) + 1}. Exit")

# Main program loop
while True:
    display_menu()
    choice = input("Select a category (or 'exit' to quit): ")

    if choice == 'exit':
        break

    try:
        choice = int(choice)
        if 1 <= choice <= len(contents):
            selected_category = list(contents.keys())[choice - 1]
            subcategories = contents[selected_category]
            print(f"Subcategories for {selected_category}:")
            for index, subcategory in enumerate(subcategories.keys(), 1):
                print(f"{index}. {subcategory}")
            print(f"{len(subcategories) + 1}. Back to Categories")
            sub_choice = input("Select a subcategory or enter to go back: ")

            if sub_choice == '':
                continue

            try:
                sub_choice = int(sub_choice)
                if 1 <= sub_choice <= len(subcategories):
                    selected_subcategory = list(subcategories.keys())[sub_choice - 1]
                    recommendations = subcategories[selected_subcategory]
                    if recommendations:
                        print("Recommendations:")
                        for recommendation in recommendations:
                            print(f" - {recommendation}")
                        append_recommendations_to_file(selected_category, selected_subcategory, recommendations)
                        print("Recommendations appended to file.")
                    else:
                        print("No recommendations available for this subcategory.")
                elif sub_choice == len(subcategories) + 1:
                    continue
                else:
                    print("Invalid subcategory choice.")
            except ValueError:
                print("Invalid input. Please enter a number for subcategory choice.")
        elif choice == len(contents) + 1:
            break
        else:
            print("Invalid category choice.")
    except ValueError:
        print("Invalid input. Please enter a number for category choice.")
```
